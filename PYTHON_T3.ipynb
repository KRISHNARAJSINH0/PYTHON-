# WAP to count words, characters and spaces in a text file

```python
space = 0
ch = 0
with open("abc.txt") as file:
    data = file.read()
    for count in data:
        ch+=1
        if count == " ":
            space+=1
    word = len(data.split())
    print("space in data :",space-1)
    print("character in data :",ch-space)
    print("word in data :",word)
    print(data)
```

# WAP that reads data from a file and calculates the percentage of vowels and consonants in the file

```python
with open("abc.txt") as file:
    vowel = 0
    consonants = 0
    data = file.read()
    for i in data:
        if i in "aeiouAEIOU":
            vowel+=1
        elif i.isalpha():
            consonants +=1 
        else:
            pass
print("no of vowel :",vowel)
print("no of consonants :",consonants)
print(f"% of vowel :{((vowel*100)/26):0.2f}")
print(f"% of consonants :{((consonants*100)/26):0.2f}")
print(data)
```

# WAP to find the longest word from a file.


```python
with open("abc.txt") as file:
    l = []
    data = file.read().split()
    mx = len(max(data,key = len))
    for word in data:
        if len(word) == mx:
            l.append(word)
    print(l)
```

# Library management sysytem

```python
class Library:
    def __init__(self):
        self.title = ""
        self.author = ""
        self.publisher = ""
    def write(self):
        self.title = input("Enter book title :")
        self.author = input("Enter author name :")
        self.publisher = input("Enter publisher name :")
    def display(self):
        print("Title :",self.title)
        print("Author :",self.author)
        print("Publisher :",self.publisher)
        print("\n")
my_list = []
ch = "y"
while (ch=="y"):
    print("1 :- Add Book")
    print("2 :- Display Book")
    x = int(input("Enter : "))
    if x==1:
        book = Library()
        book.write()
        my_list.append(book)
    elif x==2:
        for i in my_list:
            i.display()
    else:
        print("Enter valid Input")
    ch = input("for Continue Enter y, or for exit Enter n")
print("byyyy")
```

# Billing System

```python
class Billing:
    def __init__(self,n):
        self.item_code = []
        self.price = []
        self.n = n
    def add_item(self):
        for i in range(self.n):
            self.item_code.append(input(f"Enter item code {i+1} :"))
            self.price.append(int(input(f"Enter item price {i+1} :")))
    def display(self):
        for i in range(self.n):
            print(f"Item Code {i+1}:",self.item_code[i],end="\n")
            print(f"Price {i+1}:",self.price[i])
            print("\n")
    def bill(self,que):
        total = 0
        for i in range(self.n):
            print(f"price of item {i+1} with quantity {que[i]} :",self.price[i]*que[i])
        for i in range(self.n):
            total+=(self.price[i]*que[i])
        print("The total bill is :",total)
                                
            
x = int(input("Enter how many item you buy :"))
b = Billing(x)
b.add_item()
b.display()
q = []
print("Enter quantity of items :")
for i in range(x):
    q.append(int(input(f"Enter Quantity of item {i+1}:")))
b.bill(q)
```

## Bank System

```python
class Bank:
    def __init__(self,acc_no,name,balance):
        self.acc_no = acc_no
        self.name = name
        self.balance = balance
    def withdrawal(self,w):
        if self.balance < w:
            print("okat dekh apni")
        else:
            self.balance = self.balance-w
            print("you withdrawal ",w)
            print("curent balance is :",self.balance)
    def diposit(self,d):
        self.balance = self.balance + d
        print("you diposit ",d)
        print("curent balance is :",self.balance)
    def display(self):
        print("Account no :",self.acc_no)
        print("Name :",self.name)
        print("Balance :",self.balance)
b = Bank(123456790,"Devil",2300)
b.withdrawal(300)
b.diposit(500)
b.display()
```

# flashcard

```python
class flashcard:
    def __init__(self,flash,mean):
        self.flash = flash
        self.mean = mean
    def formula(self):
        return self.flash+"( "+self.mean+" )"

f = []
ch = "y"
while (ch=="y"):
    print(": Welcome to flashcard :")
    x = input("Enter short form :")
    y = input("Enter full form ")
    data = flashcard(x,y)
    f.append(data.formula())
    ch = input("Enter y for continue adding,hit any button for exit ")
print("your flash is :")
for i in f:
    print(">",i)
```

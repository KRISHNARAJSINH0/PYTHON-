```python
from bs4 import BeautifulSoup

# Read the local HTML file
with open("index.html", "r", encoding="utf-8") as file:
    html = file.read()

# Parse HTML
soup = BeautifulSoup(html, "html.parser")

# ----------------------------------------
# 1. Page Title
# ----------------------------------------
print("TITLE:")
print(soup.title.text)

# ----------------------------------------
# 2. Main Heading
# ----------------------------------------
print("\nMAIN H1:")
print(soup.find("h1").text)

# ----------------------------------------
# 3. Find by ID
# ----------------------------------------
print("\nFIND BY ID:")
element = soup.find(id="main-title")
print(element.text)

# ----------------------------------------
# 4. Find by Class
# ----------------------------------------
print("\nFIND BY CLASS:")
content = soup.find(class_="content")
print(content.text.strip())

# ----------------------------------------
# 5. Find all paragraphs
# ----------------------------------------
print("\nPARAGRAPHS:")
paragraphs = soup.find_all("p")

for p in paragraphs:
    print(p.text)

# ----------------------------------------
# 6. Find all links
# ----------------------------------------
print("\nLINKS:")

links = soup.find_all("a")

for link in links:
    print("Text :", link.text)
    print("URL  :", link.get("href"))

# ----------------------------------------
# 7. Find all images
# ----------------------------------------
print("\nIMAGES:")

images = soup.find_all("img")

for img in images:
    print(img.get("src"))

# ----------------------------------------
# 8. Find unordered list items
# ----------------------------------------
print("\nLANGUAGES:")

languages = soup.find("ul")

for item in languages.find_all("li"):
    print(item.text)

# ----------------------------------------
# 9. Find ordered list items
# ----------------------------------------
print("\nSTEPS:")

steps = soup.find("ol")

for item in steps.find_all("li"):
    print(item.text)

# ----------------------------------------
# 10. Find Student Card
# ----------------------------------------
print("\nSTUDENT CARD:")

card = soup.find("div", class_="card")

print(card.find(class_="name").text)
print(card.find(class_="age").text)
print(card.find(class_="city").text)

# ----------------------------------------
# 11. Extract Products
# ----------------------------------------
print("\nPRODUCTS:")

products = soup.find_all("div", class_="product")

for product in products:
    name = product.find(class_="product-name").text
    price = product.find(class_="price").text

    print(name, "-", price)

# ----------------------------------------
# 12. Extract Table
# ----------------------------------------
print("\nTABLE DATA:")

table = soup.find("table")

rows = table.find_all("tr")

for row in rows:
    cols = row.find_all(["th", "td"])

    data = [col.text.strip() for col in cols]

    print(data)

# ----------------------------------------
# 13. Find Inputs
# ----------------------------------------
print("\nFORM INPUTS:")

inputs = soup.find_all("input")

for inp in inputs:
    print(inp.get("name"), "-", inp.get("type"))

# ----------------------------------------
# 14. Footer
# ----------------------------------------
print("\nFOOTER:")
footer = soup.find("footer")
print(footer.text.strip())

# ----------------------------------------
# 15. CSS Selector Examples
# ----------------------------------------
print("\nCSS SELECT EXAMPLES:")

print("First product:")
print(soup.select_one(".product .product-name").text)

print("\nAll prices:")
for price in soup.select(".price"):
    print(price.text)

# ----------------------------------------
# 16. Get Attributes
# ----------------------------------------
print("\nIMAGE ATTRIBUTES:")

for img in soup.find_all("img"):
    print("src =", img.get("src"))
    print("alt =", img.get("alt"))

# ----------------------------------------
# 17. Parent Example
# ----------------------------------------
print("\nPARENT OF PRODUCT NAME:")

name = soup.find(class_="product-name")
print(name.parent.name)

# ----------------------------------------
# 18. Children Example
# ----------------------------------------
print("\nCHILDREN OF PRODUCT DIV:")

product = soup.find(class_="product")

for child in product.children:
    if child.name:
        print(child.name)

# ----------------------------------------
# 19. Next Sibling
# ----------------------------------------
print("\nNEXT PRODUCT:")

first_product = soup.find(class_="product")
next_product = first_product.find_next_sibling("div")

print(next_product.find(class_="product-name").text)

# ----------------------------------------
# 20. Pretty Print HTML
# ----------------------------------------
print("\nFORMATTED HTML:")
print(soup.prettify()[:500])  # first 500 characters
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
q = []
author = []
tag = []

a = soup.find_all(class_="quote")
for i in a:
    q.append(i.find(class_='text').text)
    author.append(i.find(class_='author').text)
    b = []
    for j in i.find_all(class_="tag"):
        b.append(j.text)
    tag.append(b)
    
data = {"Quote":q,"Author":author,"Tags":tag}
df = pd.DataFrame(data)
df
```

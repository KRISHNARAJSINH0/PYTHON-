# PIE CHART

```python
labels = ["A","B","C","D"]
size = [40,35,10,15]
colors = ["red","Yellow","Green","cyan","dark"]
```

```python
plt.pie(
    size,
    labels = labels,
    colors = colors,
    explode = [0.1,0,0,0],
    startangle = 90,
    autopct = "%1.1f%%"
)
plt.show()
```

```python
plt.pie(
    size,
    labels = labels,
    colors = colors,
    explode = [0,0.1,0.1,0.2],
    startangle = 90,
    autopct = "%1.1f%%"
)
plt.show()
```

# stackplot

```python
import matplotlib.pyplot as plt
import numpy as np
x = range(1,6)
y1 = [1,4,6,8,4]
y2 = [2,2,7,10,12]
y3 = [2,8,5,10,6]
plt.stackplot(x,y1,y2,y3,labels = ["A","B","C"])
plt.legend()
plt.show()
```

```python
%%writefile tutorial.py
import streamlit as st
st.title("This is streamlit tutorial")
st.header("Welcome to Streamlit")
st.subheader("This is subheader")
st.text("This is st.text")
st.write("This is st.write")
st.markdown("This is `st.markdown`")

code = """
def add(a,b):
    return a+b
print(add(5,6))
"""
st.code(code,language = 'python')
st.write("new line ")
st.sidebar.header("Profile setting")
name = st.sidebar.text_input("Faculty_name")
dept = st.sidebar.selectbox("Dept",['CE','CSE','IT'])
col1,col2 = st.columns([1,2])
with col1:
    st.write("Basic info")
    st.write(name,dept)
with col2:
    st.subheader("About")
    st.write("new column")
with st.expander("Subject taught"):
    st.write("Python,CN,DE")
course = st.selectbox('Course',['Python','FSD','DE'])
days = st.multiselect("Preferred Days",["MON","TUE","WED"])
lec = st.radio("Lecture No. ",[1,2,3])

from datetime import date,time
examdate = st.date_input("Exam date ",date.today())
startdate = st.time_input("Start time ",time(14,0))
file = st.file_uploader("Upload CSV ",type = ["CSV"])

# MediaData 
st.subheader("Web Image")
st.image("https://picsum.photos/600/300")
# Local image 
st.subheader("Local Image")
st.image("https://wallpapercat.com/w/full/7/9/9/607656-3840x2160-desktop-4k-sports-car-background.jpg")
st.subheader("Audio")
st.audio("funny.mp3")
st.subheader("Video")
st.video("b.mp4")
 # Create Data
import pandas as pd
data = {
    "Student ": ['A','B','C','D'],
    "Marks ": [85,92,76,24],
    "Passed ":[True,True,True,False]
}
df = pd.DataFrame(data)
st.header("st.dataframe")
st.dataframe(df)
st.header("st.table")
st.table(df)

st.subheader("JSON - Javascript Object Naotation")
st.json(data)
import time
st.info("Useful Info")
if st.button("Start Long Run"):
    progress = st.progress(0)
    with st.spinner("Progressing"):
        for i in range(100):
            time.sleep(0.03)
            progress.progress(i+1)
    st.success("Task Completed")
    st.balloons()
    st.snow()

# Charts
st.header("Matplotlib based Charts")
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,11)
y = np.random.randint(50,100,size = 10)
plt.figure(figsize = (6,5))
plt.plot(x,y,'*')
plt.xlabel("x axix")
plt.ylabel("y axis")
plt.title("Matplotlib based line Chars")
st.pyplot(plt)

st.header("Streamlit based Chart")
df1 = pd.DataFrame(
    {
        'Student ': x,
        'Marks ' : y
    }
)
st.line_chart(df1)
```

# PB 712

```python
%%writefile pb712.py
import streamlit as st
st.header("PB 712 PROGRAM")
with st.sidebar:
    country = st.selectbox("Select country",['USA','India','UK','Canada'])

totalpopulation=st.number_input("ENTER POPULATION",min_value=1)
vaccinated=st.number_input("VACCINATED PEOPLE",min_value=0)
if st.button("CALCULATE VACCINATED"):
    percentage=(vaccinated/totalpopulation)*100
    st.write(f"vaccinated for {country} is {percentage:.2f}%")
    st.progress(min(int(percentage),100))
    if percentage>=70:
        st.success("GOOD VACCINATION")
        st.snow()
    else:
        st.warning("VACCINATION POOR")
```

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y


class Slope:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2

    def find_slope(self):
        if self.p2.x - self.p1.x == 0:
            print("Slope is undefined (vertical line).")
        else:
            slope = (self.p2.y - self.p1.y) / (self.p2.x - self.p1.x)
            print("Slope between the two points is:", slope)
# Create two points
p1 = Point(2, 4)
p2 = Point(6, 10)

# Calculate slope
s = Slope(p1, p2)
s.find_slope()
````

````python
from abc import ABC, abstractmethod
import math
class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass
class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def calculate_area(self):
        return self.length * self.breadth

    # Overloading > operator
    def __gt__(self, other):
        return self.calculate_area() > other.calculate_area()
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius * self.radius
# Create Rectangle objects
r1 = Rectangle(10, 5)
r2 = Rectangle(8, 6)

print("Area of Rectangle 1:", r1.calculate_area())
print("Area of Rectangle 2:", r2.calculate_area())

# Compare rectangle areas using overloaded >
if r1 > r2:
    print("Rectangle 1 has greater area than Rectangle 2")
else:
    print("Rectangle 2 has greater or equal area than Rectangle 1")

# Create Circle object
c1 = Circle(7)
print("Area of Circle:", c1.calculate_area())

# Method Resolution Order of Circle class
print("\nMethod Resolution Order (MRO) of Circle:")
print(Circle.mro())
````

````python
%%writefile E3.py
import streamlit as st
st.title("Student Marks")
col1,col2,col3,col4,col5 = st.columns(5)
with col1:
    m1 = st.number_input("Enter Subject 1 Marks ",min_value = 0,step = 1, max_value =100)
with col2:
    m2 = st.number_input("Enter Subject 2 Marks ",min_value = 0,step = 1, max_value =100)
with col3:
    m3 = st.number_input("Enter Subject 3 Marks ",min_value = 0,step = 1, max_value =100)
with col4:
    m4 = st.number_input("Enter Subject 4 Marks ",min_value = 0,step = 1, max_value =100)
with col5:
    m5 = st.number_input("Enter Subject 5 Marks ",min_value = 0,step = 1, max_value =100)
if st.button("View"):
    total = m1+m2+m3+m4+m5
    avg = total/5
    if total >= 60 :
        div = "A Div Class"
    elif total >= 40:
        div = "B Div Class"
    else:
        div = "Fail"
    with st.expander("View Student Score"):
        st.write(f"Total Marks : {total}")
        st.write(f"AVG Marks : {avg}")
        st.write(f"Div By Marks : {div}")
````

````python
%%writefile E4.py
import streamlit as st 
st.title("BMI Calculator")
kg = st.number_input("Enter Weight In KG : ",min_value = 1,step = 1)
cm = st.number_input("Enter height In CM : ",min_value = 1,step = 1)
if st.button("Click"):
    bmi = (kg/((cm/100)**2))
    st.subheader(f"BMI Value is : {bmi}")
    if bmi < 18.5:
        st.warning("Underweight")
    elif 18.8<= bmi < 25:
        st.success("Normal weight")
    elif 25<= bmi < 30:
        st.warning("Overweight")
    else:
        st.error("obese")
````

````python
%%writefile E5.py

import streamlit as st
import matplotlib.pyplot as plt
import random

# App title
st.title("Matplotlib Integration App")

# Input: number of points
n = st.number_input(
    "Enter number of random points",
    min_value=1,
    max_value=500,
    step=1
)

# Button
if st.button("Generate Scatter Plot"):

    # Generate random x and y values
    x = [random.randint(1, 100) for _ in range(n)]
    y = [random.randint(1, 100) for _ in range(n)]

    # Create matplotlib figure
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    ax.set_xlabel("X values")
    ax.set_ylabel("Y values")
    ax.set_title("Random Scatter Plot")

    # Display plot in Streamlit
    st.pyplot(fig)
````

````python
%%writefile E6.py
import streamlit as st

# App title
st.title(" To-Do List App")

# Initialize task list in session state
if "tasks" not in st.session_state:
    st.session_state.tasks = []

# Input for new task
task = st.text_input("Enter a task")

# Add task button
if st.button("Add Task"):
    if task:
        st.session_state.tasks.append(task)
        st.success("Task added successfully!")
    else:
        st.warning("Please enter a task.")

st.write("---")

# Display tasks
st.subheader("Your Tasks")

if not st.session_state.tasks:
    st.info("No tasks added yet.")
else:
    for i, task in enumerate(st.session_state.tasks):
        if st.checkbox(task,key = i):
            st.success(f" Task completed: {task}")
````

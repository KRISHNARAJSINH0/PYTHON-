```python

# x = soup.find_all("li",{"class":"o-listicle_item"})
# n = []
# d = []
# e = []
# author = []
# for i in x:
#     name = i.find("a",{"class":"m-statement__name"}).text.strip()
#     n.append(name)
#     date = i.find("div",class_ = "m-statement__desc").text.split()
#     d.append(date)
#     fo = i.find("footer",class_ = "m-statement__footer").text.split().split("•")
#     e.append(fo[1])
#     author.append(fo[0])
soup = BeautifulSoup(html,'html.parser')
x = soup.find_all('li',{'class':'o-listicle__item'})
n=[]
d=[]
e=[]
author=[]

for i in x:
    name = i.find('a',{'class':'m-statement__name'}).text.strip()
    n.append(name)
    date = i.find("div",{'class':"m-statement__desc"}).text.strip()
    d.append(date)
    fo = i.find('footer',{'class':'m-statement__footer'}).text.strip().split('•')
    e.append(fo[1])
    author.append(fo[0])
```

```python
data = {"name":n,"date":d,"state":e,"at":author}
import pandas as pd
df = pd.DataFrame(data)
df
```

# Flipkart question

```python
with open("flipkart.html","r",encoding="utf-8") as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html,"html.parser")
```

```python
z = soup.find_all("div",class_="_3pLy-c row")
n = []
r = []
b = []
p = []
d = []
dp = []
pn = []
co = []
st = [] 
for i in z:
    name = i.find("div",class_="_4rR01T").text.strip()
    n.append(name)
    rate = i.find("div",class_="_3LWZlK").text.strip()
    r.append(rate)
    battery = i.find_all("li",class_="rgWa7D")[3]
    b.append(battery.text)
    price = i.find("div",class_="_30jeq3 _1_WHN1").text.strip()
    p.append(price)
    discount = i.find("div",class_="_3Ay6Sb")
    if discount:
        d.append(discount.text.strip())
    else:
        d.append("None")
    d_price = i.find("div",class_="_3I9_wc _27UcVY")
    if d_price:
        dp.append(d_price.text.strip())
    else:
        dp.append("None")


for i in n:
    x = i.split('(')
    pn.append(x[0])
    co.append((i.split("("))[1].split(",")[0])
    
    
    
```

```python
co
```

```python
data = {"Name":n,"Phone_Name":pn,"Color":co,"Storage":,"Rating":r,"Battery":b,"Price":p,"Discount":d,"Discount_Price":dp}
df = pd.DataFrame(data)
df
```

## networkx

### unidirected graph

```python
import networkx as nx 
```

```python
g = nx.Graph()
g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))
g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(4,5),(6,7),(5,6)])
nx.draw_networkx(g,
                width=5,
                node_size = 830,
                node_color = "skyblue",
                edge_color = "yellow")
plt.show()
# width for line width
# node_size for incress size of node
```

### directed graph

```python
g = nx.DiGraph()
g.add_node(1)
g.add_nodes_from([2,3])
g.add_nodes_from(range(4,7))
g.add_edge(1,2)
g.add_edge(1,1)
g.add_edges_from([(2,3),(4,5),(6,7),(5,6)])
nx.draw_networkx(g,
                width=2,
                node_size = 830,
                node_color = "skyblue",
                edge_color = "red")
plt.show()
# width for line width
# node_size for incress size of node
```

```python
def remove_outlier(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    IQR = q3 - q1
    low = q1 - 1.5*IQR
    high = q3 + 1.5*IQR
    return x[~((x<low) | (high<x)).any(axis=1)]

df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5, 100, 200, 300],
    'B': [10, 20, 30, 40, 50, 100, 200, 300]
})
print(remove_outlier(df))
```
